```python
# TODO: prepare your data frame so that you can work with
df_clean = df.drop_nulls()


df_clean = df_clean.with_columns(
    (pl.when(pl.col("gender") == "male").then(pl.lit(1))
    .otherwise(pl.lit(0)).alias("gender")),
    (pl.when(pl.col("outcome") == "Alive").then(pl.lit(1))
    .otherwise(pl.lit(0)).alias("outcome")),
    (pl.when(pl.col("smoker") == "Yes").then(pl.lit(1))
     .otherwise(pl.lit(0)).alias("smoker")))

df_clean.head
```
