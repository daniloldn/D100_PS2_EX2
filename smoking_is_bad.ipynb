```python
# TODO: install all the necessary packages
from pathlib import Path
import polars as pl
import statsmodels.api as sm
import pandas as pd
```

```python
# TODO: load the data using the pathlib library to create the path to your data
path = Path.cwd()/"data"/"smoking_data.csv"
df = pl.read_csv(path)
df.head()
```

```python
# TODO: prepare your data frame so that you can work with
df_clean = df.drop_nulls()


df_clean = df_clean.with_columns(
    (pl.when(pl.col("gender") == "male").then(pl.lit(1))
    .otherwise(pl.lit(0)).alias("gender")),
    (pl.when(pl.col("outcome") == "Alive").then(pl.lit(1))
    .otherwise(pl.lit(0)).alias("outcome")),
    (pl.when(pl.col("smoker") == "Yes").then(pl.lit(1))
     .otherwise(pl.lit(0)).alias("smoker")))

df_clean.head
```

```python
x= df_clean.select(["gender", "age", "salary", "smoker"]).to_pandas()
y = df_clean.select("outcome").to_pandas()


y = y.squeeze()
x = sm.add_constant(x)

model = sm.OLS(y, x).fit()

print(model.summary())
```
